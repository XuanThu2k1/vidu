## KNN Regression

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
```

```python
df = pd.read_csv("homeData.csv", index_col="Id")
df.shape
```

```python
df.head()
```

```python
df.columns
```

```python
features = ['LotArea','YearBuilt', '1stFlrSF', '2ndFlrSF','BsmtFullBath','BedroomAbvGr','TotalBsmtSF','GarageCars','SalePrice']
```

```python
data=df[features]
data.head()
```

```python
data.info()
```

```python
X=data.iloc[:,:-1]
X
```

```python
y=data.iloc[:,-1]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
knn_Reg=KNeighborsRegressor(n_neighbors = 3)
knn_Reg.fit(X_train, y_train)

y_predicted = knn_Reg.predict(X_test)
print(pd.DataFrame({"y_predicted": y_predicted, "y_test":y_test}))
```
